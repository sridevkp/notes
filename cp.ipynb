```python
#f(z) = z = z^2 + c
image = np.zeros( (720, 720), dtype = np.uint8)

def mandelbrot( x, y, max_iter, threshold ):
    c = complex( x, y ) 
    z = 0.0j

    for i in range(max_iter):
        z = z *z +c
        if( z.real *z.real +z.imag *z.imag ) > threshold:
            return i
    return max_iter
    
def draw( min, max, image, max_iter, threshold ):
    w, h = image.shape
    min_x, min_y = min
    max_x, max_y = max
    pxl_w, pxl_h = (max_x -min_x) / w, (max_y -min_y) / h

    for x in range(w):
        real = min_x + pxl_w * x
        for y in range(h):
            im = min_y + pxl_h * y
            color = mandelbrot( real, im, max_iter, threshold )
            image[x,y] = 255 - int(color * 255 / max_iter)


draw( (-2.0,-1.0), (1.0,1.0), image, 20, 4 )
imshow(image)
show()
```

```python
w = h = 200
img = np.zeros( (w, h, 3), dtype = np.uint8)

p = [(w/2,0), (0,h), (w,h)]
c= [(255,0,0), (0,255,0), (0,0,255)]

x, y = w/2, h/2

for i in range(50000):
    i = random.randint(0,2)
    nx, ny, = p[i]
    x += (nx-x) / 2
    y += (ny-y) / 2
    img[int(y), int(x)] = c[i]

imshow(img)
```
